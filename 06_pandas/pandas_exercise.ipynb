```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
diamonds = pd.read_csv('diamonds.csv',index_col=0)
```

### Task 1: Find all the diamonds with more than 2 carat and plot their price distribution in a histogram

###Task 3: Import the "Kernmerkmale Bevölkerung (Geschlecht, Deutsche/Ausländer, 5 Altersgruppen)" data set

https://www.destatis.de/DE/PresseService/Presse/Pressekonferenzen/2013/Zensus2011/zensus_pk.html

download the excel file and use the `read_excel` method in pandas (hint: useful arguments are `sheetname`, `header` and `index_col`

```python
zensus = pd.read_excel('demo_excel.xls',sheetname=2,header=3,index_col=1)
zensus.head()
```

### Task 4: Create a new DataFrame containing only the "Bundesländer" (hint: use the corresponding SATZART key)

```python
zensus_bl = zensus[ zensus['SATZART']==10 ]
zensus_bl.index = zensus_bl.NAME
zensus_bl.head()
```

### Task 5: Plot the age composition of the different Bundesländer

```python
age_cols = [x for x in zensus_bl if 'ALTER' in x ]
zensus_bl[age_cols].plot(kind='barh',stacked=True,figsize=(15,10))
```

### Task 6: ...and now relative to the population in each Bundesland

```python
relative_age = zensus_bl[age_cols].div( zensus_bl['EWZ'], axis=0 )
relative_age.plot(kind='barh', stacked=True, figsize=(15,10))
```

### Task 7: Plot pie charts of the relative age distribution in Berlin and Sachsen

```python
relative_age.loc['Berlin'].plot(kind='pie')
```

```python
relative_age.loc['Sachsen'].plot(kind='pie')
```

###Task 8: Import some health care data set from Zeit online 

https://docs.google.com/spreadsheets/d/16mD3R-K1vRlGLQyuel5tvMNpq29NX2sCkmA0tNrlxD4/edit#gid=813670830

```python
## hint: use the read_excel method of pandas
multir_keime = pd.read_excel('multiresistente_keime.xlsx', sheetname=0, index_col=0)
multir_keime.head()
```

### Task 9: Merge the two data sets (hint: "Kreisschlüssel" is the same as "AGS")

```python
merged_data = pd.merge(left=zensus, right=multir_keime, how='inner', left_index=True, right_index=True)
merged_data.head()
```

###  Task 10: Plot number of hospitalized patients per inhabitant against relative number of old people (65+)

```python
merged_data['KKH Patienten pro EW'] = merged_data['Krankenhaus- Patienten 2013']/merged_data['EWZ']
merged_data['Anteil 65+'] = merged_data['ALTER_5']/merged_data['EWZ']
merged_data.plot(kind='scatter',x='KKH Patienten pro EW',y='Anteil 65+',figsize=(10,10),xlim=[.1,.35],ylim=[.1,.35])
plt.plot([.1,.35],[.1,.35],'--r')
```

### Task 11: Compute the correlation

```python
merged_data[['KKH Patienten pro EW','Anteil 65+']].corr()
```
